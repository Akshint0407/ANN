```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tanh():
    x = np.linspace(-10,10,100)
    plt.plot(x,np.tanh(x), label = 'Tanh')
    plt.xlabel('Input')
    plt.ylabel('Activation function')
    plt.title('Tanh Activation Function')
    plt.grid()
    plt.legend()
    plt.show()
tanh()
```

```python
def sigmoid():
    x = np.linspace(-10,10,100)
    plt.plot(x,1/(1+np.exp(-x)), label = 'Sigmoid', color = 'Red')
    plt.xlabel('Input')
    plt.ylabel('Activation function')
    plt.title('Sigmoid Activation Function')
    plt.legend()
    plt.grid()
    plt.show()
sigmoid()
```

```python
def relu():
    x = np.linspace(-10,10,100)
    plt.plot(x,np.maximum(0,x), label = 'Relu', color = 'g')
    plt.xlabel('Input')
    plt.ylabel('Activation function')
    plt.title('Relu Activation Function')
    plt.grid()
    plt.legend()
    plt.show()
relu()
```

```python
def softmax():
    x = np.linspace(-10,10,100)
    exp_x = np.exp(x - np.max(x))
    plt.plot(x,exp_x, label = 'Softmax', color = 'm')
    plt.xlabel('Input')
    plt.ylabel('Activation function')
    plt.title('Softmax Activation Function')
    plt.grid()
    plt.legend()
    plt.show()
softmax()
```

```python
# Tanh
plt.subplot(2,2,1)
tanh()

# Sigmoid
plt.subplot(2,2,2)
sigmoid()

# ReLU
plt.subplot(2,2,3)
relu()

# Softmax
plt.subplot(2,2,4)
softmax()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Update functions to accept 'ax' argument (the subplot axis)
def tanh(ax):
    x = np.linspace(-10, 10, 100)
    ax.plot(x, np.tanh(x), label='Tanh')
    ax.set_title('Tanh Activation Function')
    ax.set_xlabel('Input')
    ax.set_ylabel('Output')
    ax.grid()
    ax.legend()

def sigmoid(ax):
    x = np.linspace(-10, 10, 100)
    ax.plot(x, 1 / (1 + np.exp(-x)), label='Sigmoid')
    ax.set_title('Sigmoid Activation Function')
    ax.set_xlabel('Input')
    ax.set_ylabel('Output')
    ax.grid()
    ax.legend()

def relu(ax):
    x = np.linspace(-10, 10, 100)
    ax.plot(x, np.maximum(0, x), label='ReLU')
    ax.set_title('ReLU Activation Function')
    ax.set_xlabel('Input')
    ax.set_ylabel('Output')
    ax.grid()
    ax.legend()

def softmax(ax):
    x = np.linspace(-10, 10, 100)
    ax.plot(x, np.exp(x) / np.sum(np.exp(x)), label='Softmax')
    ax.set_title('Softmax Activation Function')
    ax.set_xlabel('Input')
    ax.set_ylabel('Output')
    ax.grid()
    ax.legend()

# Create a 2x2 grid for subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Call each function and pass the corresponding subplot axes
tanh(axes[0, 0])
sigmoid(axes[0, 1])
relu(axes[1, 0])
softmax(axes[1, 1])

plt.tight_layout()
plt.show()
```
