```python
import numpy as np
```

```python
#McCulloch-Pitts Model
def mp_neuron(inputs, weights, threshold):
    net = np.dot(inputs, weights)
    if net >= threshold:
        output = 1
    else:
        output = 0
    return output
```

```python
#AndNot Neural Network
def andnot_nn(inputs):
    weights = [-2,1]
    threshold = 0
    outputs = [mp_neuron(inputs, weights, threshold) for weights in [[-1,1], [1,-1]]]
    if outputs[0] == 1 and outputs[1] == 0:
        outputs = 1
    else:
        outputs = 0
    return outputs
```

```python
#Putting Input and test Andnot Neural Network
inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
outputs = []
for i in range(inputs.shape[0]):
    y = andnot_nn(inputs[i].reshape(1,-1))
    outputs.append(y)
print("Input:", inputs)
print("Output:", outputs)
```

```python
import matplotlib.pyplot as plt
plt.plot(inputs, outputs);
```
